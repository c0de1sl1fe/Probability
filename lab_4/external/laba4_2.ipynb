### 2.3. ,     .3.   ,      (   F-)

```python
m = 3  #    
df1 = m  #      
df2 = N - m - 1  #      

RSS = sum((Y2_predicted - Y2.mean())**2)  #        
ESS = sum(errors2**2)  #     (explained sum of squares)

F_emp = RSS * (len(Y2) - m - 1) / (ESS * m)  #   F-
F_sts = sts.f.ppf(1 - alpha, df1, df2)  #    F-

if (F_emp > F_sts):  #         F-
  print("    ")  #  ,      
else:
  print("  ")  #  ,     
F_emp, F_sts  #   F-    F-
```

```python
f_test2 = results.f_test("  () = ,    () =  (1-, 0- ) = 0")  #  F-
f_test_sts = sts.f.ppf(1 - alpha, N - m - 1, m)  #    F-

print(f"F-: {f_test2}")  #   F-
if (f_test2.statistic > f_test_sts):  #     F-   
  print(" H1")  #  ,     H1
  print("  \n")
else:
  print(" H0\n")  #  ,     H0
```

```python
print("H0: b0 = 0")  #      b0
print("H1: b0   0,  ")  #      b0
Z = X.transpose().dot(X)  #     X   X
invZ = np.linalg.matrix_power(Z, -1)  #     Z
s_e_2 = (1/(N - m - 1))*sum(errors2**2)  #     

cob0 = np.sqrt(s_e_2) * np.sqrt(invZ[0, 0])  #       b0

t_emp = b[0] / cob0  #   t-
t_crit = sts.t.ppf(1 - alpha/2, N - m -1)  #    t-

if (abs(t_emp) > t_crit):  #    H0   b0
  print(" H1")  #  ,     H1
  print("   ")
else:
  print(" H0")  #  ,     H0
t_emp, t_crit  #   t-    t-
```

```python
print("H0: b1 = 0")  #      b1
print("H1: b1   0,  ")  #      b1

cob1 = np.sqrt(s_e_2) * np.sqrt(invZ[1,1])  #       b1

t_emp = b[1] / cob1  #   t-   b1
t_crit = sts.t.ppf(1 - alpha/2, N - m - 1)  #    t-

if (abs(t_emp) > t_crit):  #    H0   b1
  print(" H1")  #  ,     H1
  print("   ")
else:
  print(" H0")  #  ,     H0
t_emp, t_crit  #   t-    t-
```

```python
print("H0: b2 = 0")  #      b2
print("H1: b2   0,  ")  #      b2

cob2 = np.sqrt(s_e_2) * np.sqrt(invZ[2,2])  #       b2

t_emp = b[2] / cob2  #   t-   b2
t_crit = sts.t.ppf(1 - alpha/2, N - m - 1)  #    t-

if (abs(t_emp) > t_crit):  #    H0   b2
  print(" H1")  #  ,     H1
  print("   ")
else:
  print(" H0")  #  ,     H0
t_emp, t_crit  #   t-    t-
```

```python
print("H0: b3 = 0")  #      b3
print("H1: b3   0,  ")  #      b3

cob3 = np.sqrt(s_e_2) * np.sqrt(invZ[3,3])  #       b3

t_emp = b[3] / cob3  #   t-   b3
t_crit = sts.t.ppf(1 - alpha/2, N - m - 1)  #    t-

if (abs(t_emp) > t_crit):  #    H0   b3
  print(" H1")  #  ,     H1
  print("   ")
else:
  print(" H0")  #  ,     H0
t_emp, t_crit  #   t-    t-
```
