#### 1. Проверить, можно ли считать, что распределение случайной величины X в Северо-Кавказский ФО за весь период подчинено нормальному закону распределения. Использовать для проверки следующие три критерия: Хи-квадрат, Шапиро-Уилка и критерий Д’Агостино 

```python
x_all = np.array(data.iloc[:, 1:], dtype='float')
x_all = x_all[~np.isnan (x_all)]
a = np.mean(x_all)
s = np.var(x_all, ddof=1)
print(f'a*: {a}')
print(f'S^2: {s}')
max(x_all), min(x_all), len(x_all)
```

```python
high, bin_edges, intervals = plt.hist(x_all, edgecolor='black', bins=5, color="green", alpha=0.8, density=True)
density_abs = np.diff(bin_edges) * high * len(x_all)
print(density_abs)
plt.xticks(ticks=bin_edges)

plt.gca().xaxis.set_major_formatter('{:.2f}'.format)
plt.xlabel("Значение")
plt.ylabel('абсолютные частоты')
plt.title('Гистограмма абсолютных частот')
plt.show()
```

```python
expected_density_abs_1 = np.diff(len(x_all) * stats.norm.cdf(bin_edges, loc=a, scale=np.sqrt(s)))
expected_density_abs = expected_density_abs_1.astype(int) + 1
density_abs = density_abs.astype(int)
for i in range(len(bin_edges) - 1):
    print(f"Интервал: [{bin_edges[i]}, {bin_edges[i+1]});  Количество элементов: {density_abs[i]};  Ожидаемое количество элементов: {expected_density_abs[i]}")
```

```python
print(sum(density_abs))
print(sum(expected_density_abs))
x_all.sort()
x_all
```

```python
res = stats.chisquare(density_abs, f_exp=expected_density_abs, ddof=2)
print(f'2005-2021: {res}')
```

```python
sample = np.concatenate([np.array(x_2005), np.array(x_2010),
                             np.array(x_2015), np.array(x_2019),
                             np.array(x_2020), np.array(x_2021)])
print('Проверка гипотезы H0 с помощью теста Хи-квадрат (H0 ~ N(a,std^2)):')
result = stats.chisquare(density_abs, f_exp=expected_density_abs, ddof=2)
print(result)
if result[1] > alpha:
    print('Выборка распределена по нормальному закону => принимаем H0')
else:
    print('Выборка распределена не по нормальному закону => отвергаем H0')
```

```python
print('Проверка гипотезы H0 с помощью теста Шапиро-Уилка (H0 ~ N(a,std^2)):')
result = stats.shapiro(sample)
print(result)
if result[1] > alpha:
    print('Выборка распределена по нормальному закону => принимаем H0')
else:
    print('Выборка распределена не по нормальному закону => отвергаем H0')
```

```python
print('Проверка гипотезы H0 с помощью критерия Д’Агостино (H0 ~ N(a,std^2)):')
result = stats.normaltest(sample)
print(result)
if result[1] > alpha:
    print('Выборка распределена по нормальному закону => принимаем H0')
else:
    print('Выборка распределена не по нормальному закону => отвергаем H0')
```

```python
count = 0
for sample in sample_data:
    result = stats.shapiro(sample)
    if result[1] <= alpha:
        count += 1
print(f'Оценка вероятности совершить ошибку первого рода для {M} интервалов: {count / M}')
```

```python
count = 0
for sample in sample_data:
    result = stats.normaltest(sample)
    if result[1] <= alpha:
        count += 1
print(f'Оценка вероятности совершить ошибку первого рода для {M} интервалов: {count / M}')
```
